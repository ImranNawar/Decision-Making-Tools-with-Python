## Visualizing Categorical Data

In this notebok, we will visualize a dataset consists of `categorical data`.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Titanic Dataset
The Titanic dataset is a widely used dataset for machine learning and data analysis tasks. It contains information about passengers aboard the Titanic, including whether they survived (`survived`), their passenger class (`pclass`) etc.

In this notebook, we will focus on visualizing categorical data such as `pclass`, `sex`, and `survived` to gain insights into the distribution and relationships between these variables.

```python
titanic = sns.load_dataset('titanic')
titanic
```

**Bar Plot (Survival count by Passenger class)**

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='pclass', y='survived', data=titanic, palette='muted')
plt.title('Survival Rate by Passenger class')
plt.xlabel("Passenger class")
plt.ylabel("Survival Rate")
plt.show()
```

**Count Plot (Number of passengers per class)**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='pclass', data=titanic, palette='coolwarm')
plt.title('Number of Passegers by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

**Count Plot (Number of passengers by Gender)**

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', data=titanic, palette='Set2')
plt.title('Number of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

**Pie Chart (Survival Distribution)**

```python
survived = titanic['survived'].value_counts()
plt.figure(figsize=(8, 5))
plt.pie(survived, labels=['Did not Survive', 'Survived'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Survival Distribution')
plt.show()
```

---

**Passenger Class** (pclass)\
In the titanic dataset, `pclass` stands for Passenger class. It represents the class in which the passenger was travelling on the Titanic. This is a categorical variable with three possible values:
* First Class (upper class, wealthier passengers)
* Second Class (middle class)
* Third Class (lower class or economy)

**Survival Rate**\
The survival rate shown from 0 to 0.7 represents the proportion of passengers in each class who survived. In the Titanic dataset, the `survived` column contains values:
* 0: Survived
* 1: Survived

A survival rate of 0.7 means that 70% of the passengers in that class survived. A survival rate of 0.4 means 40% of the passengers survived, and so on.

**value_counts()**\
`value_counts()` is a pandas method that returns the counts of unique values in a Series (a column in DataFrame). It is commonly used for categorical data to see the frequency of different categories.

**autopct='%1.1f%%'**:\
Formats the percentage labels in pie charts to one decimal place
